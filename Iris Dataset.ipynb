## Explore the Iris dataset

```python
import sklearn as skl
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
names = ['sepal_length','sepal_width','petal_length','petal_width','species']
iris = pd.read_csv(url, names = names)
iris.head(10)
```

```python
iris = iris.replace(['Iris-setosa','Iris-virginica','Iris-versicolor'],['setosa','virginica','versicolor'])
iris.head(10)
```

For a biref inspection of all the bivariate scatter plots betwwen the 4 variables, along the univariate histogram for each variable, we can generate a matriz of scatter plot by group.

```python
sns.set()
scatter_matrix = sns.pairplot(iris, hue="species")
plt.title('Scatterplot Matrix')
plt.savefig('scatter_matrix.png', dpi=400)
```

Let's plot inly twoof them: Sepal length vs. Sepal width.

```python
scatter_plot = sns.scatterplot(data=iris, x=iris.sepal_length, y=iris.sepal_width, s=50, hue="species", palette="deep")
plt.title('Scatterplot - Sepal length vs. Sepal width')
plt.legend(loc="best", fontsize = "small", )
plt.savefig('scatter_plot.png', dpi=400)
```

_Data Scaling_

Now we will scale the iris dataset in order to drive all its values through 0-1

```python
iris_sc=iris
iris_sc.sepal_length = (iris.sepal_length-min(iris.sepal_length))/(max(iris.sepal_length)-min(iris.sepal_length))
iris_sc.sepal_width = (iris.sepal_width-min(iris.sepal_width))/(max(iris.sepal_width)-min(iris.sepal_width))
iris_sc.petal_width = (iris.petal_width-min(iris.petal_width))/(max(iris.petal_width)-min(iris.petal_width))
iris_sc.petal_length = (iris.petal_length-min(iris.petal_length))/(max(iris.petal_length)-min(iris.petal_length))
iris_sc.describe()
```
